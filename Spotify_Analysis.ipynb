```python
ax = sns.barplot(df["artist"].value_counts().head(), palette = "plasma")
for i in ax.containers:
    ax.bar_label(i, fmt='%d', label_type='edge', fontsize=14)
plt.title("Artists most streamed all time")
plt.ylabel("Music Count")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(df["artist_count"].value_counts().index, df["artist_count"].value_counts().values)
plt.xlabel("Number of Artists")
plt.ylabel("Songs Count")
```

```python
ax=sns.barplot(df["released_year"].value_counts().head(), palette="rainbow")
for i in ax.containers:
    ax.bar_label(i, fmt="%d", label_type='edge', fontsize=15)
plt.title("Years Most Released in")
plt.xlabel('Released Year')
plt.ylabel('Music Count')
plt.show()
```

```python
ax=sns.barplot(df["released_month"].value_counts().head(), palette="cool")
for i in ax.containers:
    ax.bar_label(i, fmt="%d", label_type='edge', fontsize=15)
plt.title("Months Most Released in")
plt.xlabel('Released Month')
plt.ylabel('Music Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax=sns.barplot(df['released_year'].value_counts().head(10),palette="viridis")
for i in ax.containers:
    ax.bar_label(i, fmt="%d", label_type='edge', fontsize=13)
plt.title("number of songs most streamed produced each year", fontsize=16)
plt.xlabel('Years')
plt.ylabel('Numbers of songs')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.histplot(df['released_year'], kde=True, palette='rainbow')
plt.title('Distribution of number of songs most streamed produced each year', fontsize=18)
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()
```
